```python
import pandas as pd
import os
from pathlib import Path
import json
```

```python
data_path = "../data/FR_simulatedCR_2025_02_14/"
files = [f for f in os.listdir(data_path) if ".json" in f]

cr = {}

for f in files:
    file_path = os.path.join(data_path, f)
    f_name = f.replace(".json", "")
    with open(file_path, "r") as f:
        cr[f_name] = json.load(f)
```

```python
def count_neg(annotations):
    count = 0
    for annot in annotations:
        neg = annot["negated"]
        count += 1 if neg else 0
    return count
```

```python
for key in cr:
    annot = cr[key]["annotations"]
    cr[key]["neg_count"] = count_neg(annot)
    print(key ,":",cr[key]["neg_count"])
```
